```python
##WAP TO FIND SUM OF FACTORS OF A  GIVEN NUMBER

n=int(input("ENTER A NUMBER : "))

sum=0
count=0
product=1
for i in range(1,n+1):
    if n % i == 0:
        print(f"Factor of {n} is {i} ")
        sum += i
        product *=i
        count += 1
    
print("SUM OF FACTORS : ", sum) 
print("product OF FACTORS : ", product)   
print("number of iterations done for getting sum of factors is : ",count) 

```

```python
## WAP TO FIND FACTORIAL OF A NUMBER


n = int(input("Enter a number: "))
fact = 1

for i in range(1, n+1):
    fact = fact * i

print(f"Factorial of a number {n} is {fact}")
```

```python
## TO FIND IF THE GIVEN NUMBER IS PRIME NUMBER OR NO

n=int(input("ENTER GIVEN NUMBER TO CHECK : "))

if n< 1:
    print(f"{n} IS NOT A PRIME NUMBER")
else:
    prime = True
    
    for i in range(2, n):
        if n % i == 0:
            prime = False
            break
    
    if prime:
        print(f"{n} is a Prime Number")
    else:
        print(f"{n} is Not a Prime")
    
```

```python
## TO FIND IF THE GIVEN NUMBER IS PRIME NUMBER OR NO

n=int(input("ENTER GIVEN NUMBER TO CHECK : "))
count=0

for i in range(1, n+1):
    if n%i==0:
        count+=1

if count ==2:
    print("it is a prime number")
else:
    print("it is not a prime number")
```

```python
## GIVEN NUMBER IS AMSTRONG NUMBER OR NOT

n=int(input("ENTER THE NUMBER : "))
og=n
c=0
s=0
while n>0:
    c = c + 1
    n = n // 10

m=og
power=c

s=0
while m>0:
    extracted=n%10
    s=s+(extracted ** power)
    m=m//10

if og ==s:
    print("THE GIVEN NUMBER IS AMSTRONG NUMBER")
else:
    print("THE GIVEN NUMBER IS NOT AMSTRONG NUMBER")

```
