```python
pwd = input("Enter the password")

if len(pwd)<8:
    print("Not Valid")
elif not (char.startswith('0123456789') for char in pwd):
    print("Not Valid")
elif not any(char.isdigit() for char in pwd):
    print("Not Valid")
elif not any(char.islower() for char in pwd):
    print("Not Valid")
elif not any(char.isupper() for char in pwd):
    print("Not Valid")
elif not any(char in '+-!@#$%^&*()?}{:;,./[]`~' for char in pwd):
    print("Not Valid")
else:
    print(f"your entered password '{pwd}' is a strong password")
```

```python
## PROGRAM TO CHECK WHETHER THE GIVEN CHARACTER IS UPPERCASE,LOWERCASE, DIGIT OR SPECIAL CHARACTER

ch=input("ENTER THE CHARACTER :")
og=ord(ch)

if og>65 and og<=90:
    print(f"THE ENTERED CHARACTER IS {ch} AND IT IS A CAPITAL LETTER")
elif og>=97 and og<=122:
    print(f"THE ENTERED CHARACTER IS {ch} AND IT IS A SMALL LETTER")
elif og>48 and og<=57:
    print(f"THE ENTERED CHARACTER IS {ch} AND IT IS A DIGIT")
else:
    print(f"THE ENTERED CHARACTER IS {ch} A SPECIAL CHARACTER")
```

```python
i = input("Enter the string: ")
s = i.lower()

print("Original String:",s)
index = len(s) - 1

while index >= 0:
    print(s[index], end='')
    index -= 1

if index==i:
    print("\nThe entered character is in pallindrome")
else:
    print("\nnot pallindrome")
```

```python
## REVERSE STRING AND FIND PALLINDROME

s=input("enter a word: ").lower()
rev=""

for ch in s:
    rev = ch + rev
print(rev)
if s==rev:
    print("pallindrome")
else:
    print("not a pallindrome")
```

```python
## CHECKING FOR ALPHA DIGIT ANS SYMBOLS IN SENTENCE AND PRINTING COUNT

s=str(input("enter a sentence"))
alpha=0
digit=0
symb=0

print(f"ENTERED SENTENCE : {s}")

for ch in s:
    if ch.isalpha():
        alpha +=1
    elif ch.isdigit():
        digit +=1
    elif ch != " ":
        symb +=1

print("alpha characters : ",alpha)
print("digit characters : ",digit)
print("symbols characters : ",symb)

if alpha>digit and alpha>symb:
    print("ALPHA CHARACTERS ARE MORE")
elif digit>alpha and digit>symb:
    print("DIGITS ARE MORE")
elif symb>digit and symb>alpha:
    print("SPECIAL CHARACTERS ARE MORE")
else:
    print("INVALID")
```

```python
## REVERSING WHOLE SENTENCE WITHOUT CHANGING A WORD

s=input("ENTER SENTENCE TO REVERSE :")

updated_s=""
word=""

for ch in s:
    if ch!= " ":
        word += ch
    else:
        updated_s=word + " " + updated_s
        word= " "

updated_s=word +updated_s

print(updated_s)
```

```python
## FINDING MAX LENGTH OF WORD IN SENTENCE

s=input("enter sentence") + " "
word=""
max_len=0

for ch in s:
    if ch!= " ":
        word += ch
    else:
        if len(word) > max_len:
            max_len=len(word)
        word=""
print(word,max_len)
```
