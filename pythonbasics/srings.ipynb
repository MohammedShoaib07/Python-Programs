```python
pwd = input("Enter the password")

if len(pwd)<8:
    print("Not Valid")
elif not (char.startswith('0123456789') for char in pwd):
    print("Not Valid")
elif not any(char.isdigit() for char in pwd):
    print("Not Valid")
elif not any(char.islower() for char in pwd):
    print("Not Valid")
elif not any(char.isupper() for char in pwd):
    print("Not Valid")
elif not any(char in '+-!@#$%^&*()?}{:;,./[]`~' for char in pwd):
    print("Not Valid")
else:
    print(f"your entered password '{pwd}' is a strong password")
```

```python
## PROGRAM TO CHECK WHETHER THE GIVEN CHARACTER IS UPPERCASE,LOWERCASE, DIGIT OR SPECIAL CHARACTER

ch=input("ENTER THE CHARACTER :")
og=ord(ch)

if og>65 and og<=90:
    print(f"THE ENTERED CHARACTER IS {ch} AND IT IS A CAPITAL LETTER")
elif og>=97 and og<=122:
    print(f"THE ENTERED CHARACTER IS {ch} AND IT IS A SMALL LETTER")
elif og>48 and og<=57:
    print(f"THE ENTERED CHARACTER IS {ch} AND IT IS A DIGIT")
else:
    print(f"THE ENTERED CHARACTER IS {ch} A SPECIAL CHARACTER")
```

```python
## PROGRAM TO CHECK WHETHER THE GIVEN CHARACTER IS UPPERCASE,LOWERCASE, DIGIT OR SPECIAL CHARACTER

ch=input("ENTER THE CHARACTER :")

if len(ch)>1:
    print("ENTER ONLY 1 CHARACTER")
elif ch.isupper():
    print("UPPER CASE CHARACTER")
elif ch.islower():
    print("LOWER CASE CHARACTER")
elif ch.isdigit():
    print("DIGIT")
else:
    print("SPECIAL SYMBOL")
```

```python
i = input("Enter the string: ")
s = i.lower()

print("Original String:",s)
index = len(s) - 1

while index >= 0:
    print(s[index], end='')
    index -= 1

if index==i:
    print("\nThe entered character is in pallindrome")
else:
    print("\nnot pallindrome")
```
