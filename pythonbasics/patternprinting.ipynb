```python
n=int(input("ENTER THR NUMBER :"))
row=n
star=1
space=row-1

for i in range(row):    
    for j in range(space):
        print("  ",end="")
    for k in range(star):
        print("* ",end="")
    print()
    star+=2
    space-=1    
```
