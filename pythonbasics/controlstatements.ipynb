```python
## IF ELSE ELIF STATEMENTS

num=int(input("Enter number : "))

if num>0:
    print("THE GIVEN NUMBER IS POSITIVE")
elif num<0:
    print("THE GIVEN NUMBER IS NEGATIVE")
else:
    print("THE GIVEN NUMBER IS ZERO")
```

```python
## ROCK PAPER SCISSOR USING ELIF STATEMENT (1)

op=str(input("Enter (rock, paper, scissor) : "))

if op=='rock':
    print("paper - I won !")
elif op=='paper':
    print("scissor - I won !")
else:
    print("rock - I won !")
```

```python
## TO CHECK WHETHER THE THE GIVEN YEAR IS LEAP YEAR OR NOT

year=int(input("Enter the year"))

if year%4==0:
    print(year,"is a leap year")
else:
    print(year,"is not a leap year")
```

```python
## ROCK PAPER SCISSOR USING MATCH STATEMENT

op=str(input("Enter (rock, paper, scissor) : "))

match op:
    case 'rock':
        print("paper - I won !")
    case 'paper':
        print("scissor - I won !")
    case 'scissor':
        print("rock - I won !")
```

```python
##  PROGRAM TO CHECK MIDDLE NUMBER OF 3 DIGIT VALUE AND THAT NUMBER IS DIVISIBLE BY 3 OR NO

num=int(input("Enter three digit number : "))

middle_num = (num // 10) % 10
print("middle number is : ", middle_num)

if middle_num%3==0:
    print(middle_num,"is divisible by 3")
else:
    print(middle_num,"is not divisible by 3")
```

```python
## TAKING 3 DIGIT NUMBER AND USING POSITION PRINTING IT ON SCREEN

num=int(input("Enter three digit number : "))
choice=str(input("Enter the pos of number : "))

if choice=='1':
    print(int(num/100))
elif choice=='2':
    print(int((num // 10)%10))
elif choice=='3':
    print(int(num%10))
else:
    print("-1")
```

```python
## program to take only 4 digit number and find the sum of 1st and last digit

num=int(input("Enter four digit number : "))
if num>999 and num<=9999:
    sum=(num//1000)+(num%10)
    print("the sum of 1st and last digit is : ",sum)
else:
    print("enter only 4 digit number")


```

```python
## ADDING OF MIDDLE NUMBERS FOR A4 DIGIT NUMBER

num=int(input("Enter four digit number : "))

if num>999 and num<=9999:
    sum=((num//100)%10)+((num//10)%10)
    print("the sum of 1st middle num and 2nd middle num is : ",sum)
else:
    print("enter only 4 digit number")
```
