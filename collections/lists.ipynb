```python
# LIST FUNCTIONS

marks =[2,34,34,32,32,3,78]
details = [20,'shoaib',98.7,True]

marks.append(26) # we can add element in lists 
details.append(False)
print("append :",marks,"\n",details)

marks.pop(3)  # removing an ele from lists using index 
details.pop(0)
print("pop :",marks,"\n",details)

marks.insert(5,99)  # inserting an ele in  lists 
details.insert(3,True)
print("insert :",marks,"\n",details)

marks.remove(2)  # removing an ele from lists 
details.remove(True)
print("remove :",marks,"\n",details)

marks.reverse()  # reverseing  an ele from lists 
details.reverse()
print("reverse :",marks,"\n",details)

# marks.sort()  # removing an ele from lists 
# # details.sort()
# print("sort :",marks,"\n")

print(marks.count(34))

print(marks.index(34))
print(sorted(marks))

```

```python
# FINDING MAXIMUM VALUE FROM LIST

lst=list(map(int,input("enter list ele:").split()))
max=0
for ele in lst:
    if ele>max:
        max=ele
print("maximum value : ",max)
```

```python
# SORTING EVEN AND ODD FROM LIST AND STORING IN DIFFERENT LISTS

lst=list(map(int,input("enter list ele:").split()))
lst_even=[]
lst_odd=[]

for ele in lst:
    if ele%2 ==0:
        lst_even.append(ele)
    else:
        lst_odd.append(ele)
print("even :",lst_even)
print("odd :",lst_odd)
```
