```python
d={
    1:"AIML",
    "name":"Shoaib",
    "list":[1,2,3,4],
    "tuple":(1,2,3,4),
    "dict":{"sub1":"pyhton","sub2":"java"}
}

for key in d.keys():
    print(key,"=>",d[key])

print("-------------------")

for value in d.values():
    print(value)

print("-------------------")

for key,value in d.items():
    print(key,"=>",value)
```

```python
ip=int(input("ENTER NUMBER OF ITEMS:"))
d={}

for i in range(ip):
    key=input(f"ENTER KEY{i+1}:")
    value=input(f"ENTER VALUE{i+1}:")
    d[key]=value
print(d)
```
