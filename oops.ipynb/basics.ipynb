```python
import random as r

class MyBank:
    def __init__(self,accno,name,balance):
        self.accno=accno
        self.name=name
        self.balance=balance
    
    def createcustomer(self):
        self.accno=r.randint(10000,60000)
        self.name=input("enter customer name :")
        self.balance=int(input("enter initial balance :"))

    def displaydetails(self):
        print(f"\ncustomer details\n")
        print(f"customer name : {self.name}")
        print(f"account number : {self.accno}")
        print(f"account balance : {self.balance}\n")

    def withdraw(self,amount):
        if amount > self.balance:
            print("INSUFFICIENT BALANCE\n")
        else:
            self.balance -=amount
            print(f"{amount} withdraw successful\n Available balance : {self.balance}\n")

    def deposit(self,amount):
        self.balance +=amount
        print(f"{amount} successfully deposited \n Available balance : {self.balance}")

customer=MyBank(12345,"shoaib",1000)

while True:
    print("1.create\n2.display\n3.withdraw\n4.deposit\n5.exit\n")
    ch=int(input("ENTER YOUR CHOICE :"))

    if ch==1:
        customer.createcustomer()
    elif ch==2:
        customer.displaydetails()
    elif ch==3:
        amount=int(input("ENTER AMOUNT TO WITHDRAW :"))
        customer.withdraw(amount)
    elif ch==4:
        amount=int(input("ENTER AMOUNT TO deposit :"))
        customer.deposit(amount)
    elif ch==5:
        print("EXITED !!!")
        break
    else:
        print("INVALID CHOICE")
```
