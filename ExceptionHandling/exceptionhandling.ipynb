```python
try:
    a=int(input("enter int a : "))
    b=int(input("enter int b : "))
    d=a/b
    print(f"division of {a} and {b} is :",d)
except ZeroDivisionError:
    print("dont divide by zero !!!")
except ValueError:
    print("please enter only integer values !!!")
else:
    print("try block executed successfully without any errors")
finally:
    print("I don't care if try block having error or not i will execute!!!")
print("end")
```
