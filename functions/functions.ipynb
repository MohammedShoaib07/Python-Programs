```python
def isprime(n):
    if n<=1:
        return False
    count=0
    for i in range(1,n+1):
        if n%1==0:
            count+=1
    return True if count == 2 else False

start=int(input("enter start value: "))
end=int(input("enter end value: "))
lst=[]

for i in range(start,end+1):
    if isprime(i)==True:
        lst.append(i)

print(lst)
print(len(lst))
```
