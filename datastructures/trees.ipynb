```python
class Graph:
    def __init(self,nodes):
        self.nodes = nodes
        self.matrix = [[0]*nodes for _ in range(nodes)]
            
    def addNode(self,i,j):
        self.matrix[i][j] = 1
        self.matrix[j][i] = 1

    def printMatrix(self):
        for row in self.matrix:
            print(*row)


def Main():
    noofNodes = int(input())
    noofEdges = int(input())  

    obj = Graph(noofNodes)

    for _ in range(noofEdges):
        i,j = map(int,input().split())
        obj.addNode(i-1,j-1)
    obj.printMatrix()
Main()
```
