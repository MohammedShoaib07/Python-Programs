```python
class Node:
    def __init__(self,d):
        self.data=d
        self.next=none

class Main:
    @staticmethod
    def main():
        lst = [10,20,30]
        head=None
        node=Node(lst[0])
        head=node
        print(f"[{head.data} | {head.next}]at {head}")
        node=Node(lst[1])
        head.next=node
        print(f"[{head.data} | {head.next}]at {head}")
        print(f"[{head.next.data} | {head.next.next}] at {head.next}")
        node=Node(lst[2])
        head.next.next=node
        print(f"[{head.next.data} | {head.next.next}] at {head.next}")
        print(f"[{head.next.next.data} | {head.next.next.next}] at {head.next.next}")
```
