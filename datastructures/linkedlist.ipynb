```python
class Node:
    def __init__(self,data):
        self.data=data
        self.next=None
    
class LinkedListOperation:
    def __init__(self):
        self.head=None
        
    def addNode(self,data):
        node = Node(data)
        if self.head==None:
            self.head=node
        else:
            temp=self.head
            while temp.next!=None:
                temp=temp.next
            temp.next=node
        
        
    def addAt(self,position,value):
        node=Node(value)
        if position==1:
            node.next=self.head
            self.head=node
        else:
            temp = self.head
            for _ in range(1,position-1):
                temp = temp.next
            
            node.next = temp.next
            temp.next = node
                
    

    def printLL(self):
        temp=self.head
        while temp !=None:
            print(temp.data,end=" ")
            temp=temp.next
        

    
    
class Main:
    @staticmethod
    def main():
        length=int(input())
        lst=list(map(int,input().split()))
        position=int(input())
        value=int(input())
        
        obj=LinkedListOperation()
        
        for ele in lst:
            obj.addNode(ele)
        obj.addAt(position,value)
        obj.printLL()
Main.main()
```

```python
from collections import deque
class StackOperations:
    def __init__(self,capacity):
        self.stack = []
        self.capacity = capacity
        
    def isEmpty(self):
        return True if len(self.stack) == 0 else False
        
    def push(self,val):
        if len(self.stack) == self.capacity:
            print("Stack OverFlow")
        else:
            self.stack.append(val)
        
    def pop(self):
        if  len(self.stack) == 0:
            print("Stack Underflow")
        else:
            self.stack.pop()
        
    def peek(self):
        return self.stack[-1]
        
   
#------------------------------------------------------
c = 5
s= StackOperations(c)
print(s.isEmpty())
s.pop()


            
```
